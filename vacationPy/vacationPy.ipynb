```python
 # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

```python
csvfile = '../weatherPy/weather_data_cities.csv'
weather_df = pd.read_csv(csvfile)
weather_df.head()
```

```python
fig = gmaps.figure()

locations = weather_df[['Latitude', 'Longitude']].astype(float)
humidityvalue = weather_df['Humidity'].astype(float)

heat_layer = gmaps.heatmap_layer(locations, weights = humidityvalue, 
                                 dissipating=False, max_intensity=weather_df['Humidity'].max(), 
                                 point_radius = 2)

fig.add_layer(heat_layer)

fig
```

```python
narrow_df = weather_df.loc[(weather_df['Max Temperature (F)'].astype(int) < 80) | (weather_df['Max Temperature (F)'].astype(int) > 70), :]
narrower_df = narrow_df.loc[(narrow_df['Wind Speed'] < 10), :]
hotel_df = narrower_df.loc[(narrower_df['Cloudiness'] == 0), :].astype(object)
hotel_df['Hotel Name'] = ''
hotel_df.head()
#len(hotel_df)
```

```python
lat = []
lng = []

for i in range(len(hotel_df)):
    lat.append(hotel_df.iloc[i, 1])
    lng.append(hotel_df.iloc[i, 2])
    

print(lat)
print(lng)
urls = []

#url = url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=-33.8670522%2C151.1957362&radius=1500&type=restaurant&keyword=cruise&key=YOUR_API_KEY"

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

for i in range(len(lat)):
    urls.append(f'{base_url}?location={lat[i]}%2C{lng[i]}')

print(urls)

parameters = { 
            'type' : 'lodging',
            'radius': 5000,
            'key' : g_key
}
hotel_names = []

for url in urls:
    response = requests.get(url, parameters)
    data = response.json()
    hotel_names.append(data['results'][0]['name'])


   
```

```python
hotel_df = hotel_df.iloc[:-1 , :]
hotel_df = hotel_df.iloc[:-1 , :]
hotel_df = hotel_df.iloc[:-1 , :]

hotel_df['Hotel Name'] = hotel_names
hotel_df.head()

```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{County}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
markers = gmaps.marker_layer(locations, 
                            info_box_content = [f'{hotel}' for hotel in hotel_info])

fig.add_layer(markers)
fig
```
